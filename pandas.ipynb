## Merge, Concatenate and Join

```python
# Simply Merge
df1  = pd.DataFrame({
    "EID": [1,2,3,4,5],
    "Names": ["Asad", "Raza", "Rashid", "Meesam", "Majid"],
    "Age": [32,45,25,27,30]
})
df2 =  pd.DataFrame({
    "EID": [1,2,3,4,5],
    "Salary": [45000, 50000, 40000, 43000,47000]
})
dataset = pd.merge(df1,df2)
dataset
```

```python
# Left Merge
dataset = pd.merge(left = df1, right = df2, on = "EID", how = "left")
dataset
```

```python
# Right Merge
dataset = pd.merge(left = df1, right = df2, on = "EID", how = "right")
dataset
```

```python
# Join or Concatenate
dataset = pd.concat([df1, df2], axis = 1)
dataset
```
